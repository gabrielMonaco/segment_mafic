```python
im = '/content/drive/MyDrive/MEV - metamaficas/Projeto_Segmentação_Rocha_Máfica/Ground_truth/images/AT-Z-01-32.tif'

imagem_original = Image.open(im)
imagem_original.size
```

```python
import numpy as np

# Tamanho da imagem original
altura_original = 1024
largura_original = 861

# Tamanho dos patches
tamanho_patch = 256

# Número de patches em cada dimensão
num_patches_vertical = int(np.ceil(altura_original / tamanho_patch))
num_patches_horizontal = int(np.ceil(largura_original / tamanho_patch))

# Tamanho total após o padding
altura_total = num_patches_vertical * tamanho_patch
largura_total = num_patches_horizontal * tamanho_patch

# Padding vertical e horizontal
padding_vertical = altura_total - altura_original
padding_horizontal = largura_total - largura_original

# Número de imagens que cabem em cada eixo
num_imagens_vertical = num_patches_vertical
num_imagens_horizontal = num_patches_horizontal

print("Padding vertical:", padding_vertical)
print("Padding horizontal:", padding_horizontal)
print("Número de imagens no eixo vertical:", num_imagens_vertical)
print("Número de imagens no eixo horizontal:", num_imagens_horizontal)

```
