```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
# Diretórios de origem e destino das imagens
diretorio_origem_imagens = 'caminho/para/pasta/imagens'
diretorio_origem_mascaras = 'caminho/para/pasta/mascaras'
diretorio_destino_imagens_treinamento = 'caminho/para/pasta/imagem_treino'
diretorio_destino_imagens_teste = 'caminho/para/pasta/imagem_teste'
diretorio_destino_mascaras_treinamento = 'caminho/para/pasta/mascara_treino'
diretorio_destino_mascaras_teste = 'caminho/para/pasta/mascara_teste'

# Lê o nome dos arquivos de imagens e máscaras
nomes_imagens = os.listdir(diretorio_origem_imagens)
nomes_mascaras = os.listdir(diretorio_origem_mascaras)
```

```python
# Divisão dos dados em treinamento e teste
proporcao_teste = 0.2  # Proporção de 20% para teste (você pode ajustar conforme necessário)
nomes_imagens_treinamento, nomes_imagens_teste, nomes_mascaras_treinamento, nomes_mascaras_teste = train_test_split(
    nomes_imagens, nomes_mascaras, test_size=proporcao_teste, random_state=42)

# Função auxiliar para mover arquivos
def mover_arquivos(origem, destino, nomes_arquivos):
    for nome_arquivo in nomes_arquivos:
        origem_arquivo = os.path.join(origem, nome_arquivo)
        destino_arquivo = os.path.join(destino, nome_arquivo)
        shutil.move(origem_arquivo, destino_arquivo)

# Move as imagens para os diretórios de treinamento e teste
mover_arquivos(diretorio_origem_imagens, diretorio_destino_imagens_treinamento, nomes_imagens_treinamento)
mover_arquivos(diretorio_origem_imagens, diretorio_destino_imagens_teste, nomes_imagens_teste)

# Move as máscaras para os diretórios de treinamento e teste
mover_arquivos(diretorio_origem_mascaras, diretorio_destino_mascaras_treinamento, nomes_mascaras_treinamento)
mover_arquivos(diretorio_origem_mascaras, diretorio_destino_mascaras_teste, nomes_mascaras_teste)
```
